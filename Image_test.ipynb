```python
import numpy as np
import tensorflow as tf
import pickle
import cv2
import os
import pandas as pd 
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow import keras
from gtts import gTTS
from playsound import playsound 
```

```python
model_path = "cnn.h5"
model = keras.models.load_model(model_path)
```

```python
img = cv2.imread("test/98.jpg")

plt.imshow(img)
plt.show()
```

```python
img = np.asarray(img)
img = cv2.resize(img, (128, 128))
print(img.shape)
img = img/255
img = img.reshape(1, 128, 128, 3)
```

```python
pred = model.predict(img)
result = pred.argmax()
result
```

```python
classNo = result

def getCalssName(classNo):
  if   classNo == 0: return ' '
  elif classNo == 1: return ' '
    
state=getCalssName(classNo)
print(state)
```

```python
text = (state)
language = 'bn'
```

```python
output = gTTS(text=text, lang=language,slow=True)
output.save("output.mp3")
```

```python
os.startfile("output.mp3") 
```
